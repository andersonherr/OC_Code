```python
#COMPLETE____OPS CLOCK GEN
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

# Function to convert ISO formatted datetime string to hours and days
def iso_to_hours_days(iso_str):
    dt = datetime.fromisoformat(iso_str.replace('Z', ''))
    hours = dt.hour + dt.minute / 60.0 + dt.second / 3600.0
    days = (dt - datetime(dt.year, dt.month, 1)).days
    return hours, days

# Dictionary to store Figure objects
figures = {}

# Iterate through dictionary keys
for name, dtgs in data_dict.items():
    # Create a polar heatmap for each key
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

    # Define the number of rings and wedges
    num_rings = 30
    num_wedges = 24

    # Set the radial and angular values
    r = np.arange(0, num_rings + 1, 1)
    theta = np.linspace(0, 2*np.pi, num_wedges, endpoint=False)

    # Plot the cells based on the data
    for dtg in dtgs:
        hours, days = iso_to_hours_days(dtg)
        ring_idx = int(days % num_rings)
        # Calculate the starting and ending angles for the entire hour
        start_angle = np.radians(-(int(hours) + 12) % 24 * 15)   # Adjusted to make it clockwise
        end_angle = np.radians(-(int(hours) + 13) % 24 * 15)  # Shifted to make it clockwise
        # Plot a filled wedge for the entire hour
        color = 'red'
        ax.fill_between([start_angle, end_angle], r[ring_idx], r[ring_idx + 1], color=color, alpha=0.7)

    ax.set_xticks(theta)
#     ax.set_xticklabels([str((i + 12) % 24) for i in range(24)], rotation=45, ha='right')
    ax.set_xticklabels([str((i + 13) % 24) for i in reversed(range(24))], rotation=45, ha='center')
    ax.set_yticks(r[:-1])
    ax.set_yticklabels([])

    # Set the grid
    ax.grid(True)
    # Shift the entire plot 90 degrees to the left
    ax.set_theta_offset(-np.pi / 2)

    # Set the title for each subplot
    ax.set_title(name, loc='left', va='top', fontsize=14)

    # Save the Figure object in the dictionary
    figures[name] = fig

    # Save the polar heatmap for each key as an image
    plt.savefig(f"{name}_clock.png", format='png', dpi=300, bbox_inches='tight')

    # Close the current figure to free up resources
# Now, you can use the saved Figure objects in future code
# Example: Display the saved Figure for 'John'
```

```python
#COMPLETE______MAP GEN
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# ... (your existing data_dict and eob_dicts)

def generate_plot(data_dict, output_filename):
    # Extracting latitude and longitude values
    latitudes = []
    longitudes = []
    for facility_key, facility_value in data_dict.items():
        latitudes.append(float(facility_value[0]))
        longitudes.append(float(facility_value[1]))

    # Create a scatter plot on a map
    fig, ax = plt.subplots()
    m = Basemap(projection='mill', llcrnrlat=min(latitudes) - 5, urcrnrlat=max(latitudes) + 5,
                llcrnrlon=min(longitudes) - 5, urcrnrlon=max(longitudes) + 5, resolution='l')
    m.drawmapboundary(fill_color='darkgrey')
    m.drawcoastlines(linewidth=0.5)
    m.drawcountries(linewidth=0.5)

    # Scatter plot
    x, y = m(longitudes, latitudes)
    m.scatter(x, y, color='red', marker='o', label='Facilities')

    # Add labels for each key with black background and white text on a single line
    for facility_key, facility_value in data_dict.items():
        label = f'{facility_key}'
        x, y = m(float(facility_value[1]), float(facility_value[0]) + 1.5)
        plt.text(x, y, label, fontsize=6, ha='center', va='bottom', bbox=dict(facecolor='black', edgecolor='none', boxstyle='round,pad=0.1', alpha=0.7), color='white')
    plt.title(facility_key)
    # Save the plot as a PNG file
    plt.savefig(output_filename, dpi=300, bbox_inches='tight', pad_inches=0)

    plt.close()

# Example usage:
maps_dict = {}

for key, data_dict in eob_dicts.items():
    output_filename = f'plotted_{key}.png'
    generate_plot(data_dict, output_filename)

    # Open the saved image and store the Figure object in the dictionary
    img = plt.imread(output_filename)
    fig, ax = plt.subplots()
    ax.imshow(img)
    ax.axis('off')
    maps_dict[key] = fig

# Now, you can use the saved Figure objects in future code
# Example: Display the saved Figure for 'A123B'
maps_dict['A123B'].show()
```
